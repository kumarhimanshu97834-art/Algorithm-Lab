```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
x=print(func4([1, 2, 3, 4], [2, 4, 6, 8]))
#this function is counting how many elements of arr a are also prsent in arr b
start_time = timeit.default_timer()
func4([1, 2, 3, 4], [2, 4, 6, 8])
print('>> elapsed time', (timeit.default_timer() - start_time) * 1000, 'ms')
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
x=print(func5(16))
#this function gives the square root of the number
start_time = timeit.default_timer()
func5(16)
print('>> elapsed time', (timeit.default_timer() - start_time) * 1000, 'ms')
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
x= print(func8([3,1,2,7,8,6]))
#sorts the array A in ascending order using recursion
start_time = timeit.default_timer()
func8([3,1,2,7,8,6])
print('>> elapsed time', (timeit.default_timer() - start_time) * 1000, 'ms')
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
x=print(func12([1, 2, 3, 4, 5], 2))
# this fun  rotates an array to the right by k positions.
start_time = timeit.default_timer()
func12([1, 2, 3, 4, 5], 2)
print('>> elapsed time', (timeit.default_timer() - start_time) * 1000, 'ms')
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
    minimum = arr[0]
    maximum = arr[0]

    for i in range(1, len(arr)):
        if arr[i] < minimum:
            minimum = arr[i]
        if arr[i] > maximum:
            maximum = arr[i]

    return minimum, maximum
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
arr = [1, 3, 5, 7, 9,15,16,14,45]
print(Binary_Search(arr, 5))
import time
start_time = time.perf_counter()
Binary_Search(arr, 5)
# Replace with your function and input
end_time = time.perf_counter()

# Calculate time in milliseconds
execution_time_ms = (end_time - start_time) * 1000
print(f"Execution time: {execution_time_ms:.2f}ms")
```
